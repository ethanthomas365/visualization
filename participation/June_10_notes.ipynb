```python
# first make our sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

# now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1, color="indigo")
ax.plot(x, y2, color="red")

```

```python
#adding the error bar information and re plotting.
y2_sd = np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")
ax.errorbar(x,       # our x values
            y2,      # our y values
            yerr=y2_sd,
            fmt="none")
```

```python
#copying code from above and then playing around/customizing the error bar. 
fig, ax = plt.subplots(figsize=(7, 3))

ax.plot(x, y2, color="red")

ax.errorbar(x, y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4)

```

```python
#customizing and adding intervals to the error bars.
fig, ax = plt.subplots(figsize=(7, 3))

ax.plot(x, y2, color="red")

ax.errorbar(x, y2,
            yerr=y2_sd,
            fmt="none",
            ecolor="indigo",
            elinewidth=4,
            capsize=6,
            capthick=4,
            errorevery=2)
```

```python
#Adding images to plots, following along from the slides here. 
#first library stuff
from PIL import Image # to open images
import requests # to get images from URLs
from io import BytesIO # to store images
```

```python
url = "https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png"
headers = {"User-Agent": "Mozilla/5.0",
           "Accept": "image/*,*/*;q=0.8",
           "Referer": "https://en.wikipedia.org/"}
response = requests.get(url, headers=headers)
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

ax_image = fig.add_axes([0.1,  # x coordinate (ON FIGURE, NOT AXES)
                         0.11, # y coordinate (ON FIGURE, NOT AXES)
                         0.15, # image width
                         0.35] # image height
                        )
ax_image.imshow(image)
```
