REVIEW ON MATPLOTLIB

```python
import numpy as np
import matplotlib.pyplot as plt

# Recreate the sample data from the previous notebook
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
# Review: create a Figure and one Axes inside it
# fig = the overall container; ax = the plot inside it
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
fig.show()
```

```python
# Review: a Figure can hold MULTIPLE axes (subplots)
# ncols=2 creates two side-by-side axes in one figure
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 3))

ax1.scatter(x, y)
ax1.set_title('Axes 1 (scatter)')

ax2.plot(x, y, color='orange')
ax2.set_title('Axes 2 (line)')

fig.suptitle('One Figure, Two Axes')  # suptitle sets a title for the whole figure
fig.tight_layout()
plt.show()
```

Activity: How do we decide the right visualization. Look at both the visualizations related to gun violence in the states and discuss what information can be learned and the neutrality of the visualization. 

**Visualization #1 — Periscopic: *US Gun Killings in 2018***
- An animated visualization showing expected **years of life lost** to gun violence
- Emphasizes **emotion**: each arc represents a stolen life; the visual encodes a sense of grief and loss
- Has been criticized as *"actively shaping data to support a cause"* in this case, highlighting lack of gun control
- Won several year-end information visualization awards

**Visualization #2 — Washington Post / Wonkblog: *The Era of "Active Shooters"***
- A clean, 2D bar chart showing number of active shooter incidents per year, 2000–2015 (source: FBI)
- Intended to present *"a deliberately neutral emotional field, a blank page in effect, upon which viewers are more free to choose their own response to the information"*

Design elements that contribute to the neutral feel:
- Simple 2D bar chart format (familiar, low cognitive load)
- No animation or dramatic colour palette
- Clean white background, no emotional framing
- Axis labels directly on bars
- Explicit data source cited at the bottom (FBI)

Note, it is almost impossible for data and data visualization to be absolutely neutral.
